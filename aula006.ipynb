```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("pagamentos.csv", sep=";", encoding="utf-8")

print(df)

def grafico_barras():
    plt.figure()
    plt.bar(df["Nome"], df["Salario Liquido (R$)"])
    plt.xticks(rotation = 45)
    plt.title("Salario Liquido do Professor")
    plt.xlabel("Professor")
    plt.ylabel("Salário Liquido")
    plt.tight_layout()
    plt.show()

def grafico_linhas():
    plt.figure()
    plt.plot(df["Nome"], df["Salario Liquido (R$)"])
    plt.xticks(rotation=45)
    plt.title("Evolução dos Salários")
    plt.xlabel("Professores")
    plt.ylabel("Salário (R$)")
    plt.tight_layout()
    plt.show()

def grafico_pizza():
    plt.figure()
    plt.pie(df["Salario Liquido (R$)"], labels=df["Nome"], autopct='%1.1f%%')
    plt.title("Distribuição dos Salários")
    plt.show()

def grafico_histograma():
    plt.figure()
    plt.hist(df["Salario Liquido (R$)"], bins=5)
    plt.title("Distribuição de Salários")
    plt.xlabel("Faixa Salarial")
    plt.ylabel("Quantidade")
    plt.show()

def grafico_dispersao():
    plt.figure()
    plt.scatter(df["Carga Horaria (h/mes)"], df["Salario Liquido (R$)"])
    plt.title("Carga Horária vs Salário")
    plt.xlabel("Carga Horária")
    plt.ylabel("Salário")
    plt.show()

def menu():
    while True:
        print("\n ===============TIPOS DE GRÁFICOS===============")
        print("1 -  Gráfico de Barra")
        print("2 -  Gráfico de Linhas")
        print("3 -  Gráfico de Pizza")
        print("4 -  Gráfico de Histograma")
        print("5 -  Gráfico de Dispersão")
        print("6 -  Sair")

        option = input("Escolha o numero do gráfico")

        if option == "1":
            grafico_barras()
        elif option == "2":
            grafico_linhas()
        elif option == "3":
            grafico_pizza()
        elif option == "4":
            grafico_histograma()
        elif option == "5":
            grafico_dispersao()
        elif option == "6":
            print("Encerrando...")
            break
        else:
            print("Opção Inválida")
menu()
          
```
