```python
import matplotlib.pyplot as plt

meses = ["Jan", "Fev", "Mar", "Abr", "Mai", "Jun","Jul", "Ago", "Set", "Out", "Nov", "Dez"]
vendas = [1000, 2, 10000, 5, 0, 3, 20000, 15000, 15500, 15550, 67, 6]

plt.plot(meses, vendas, marker='o')
plt.xlabel('Mês')
plt.ylabel('Vendas (em milhares)')
plt.title('Crescimento das vendas no Ano 2025')
plt.grid(True) 
plt.show()
```

```python
import matplotlib.pyplot as plt

cidades = ['Rio de Janeiro', 'Salvador', 'Belo Horizonte']
temperatura = [42, 45, 21]

formatacao = plt.bar(cidades, temperatura, color='lime')
plt.bar_label (formatacao, labels=temperatura, padding=3, fontsize=10, fontweight='bold')
plt.xlabel('Cidades')
plt.ylabel('Temperatura Média (ºC)')
plt.title('Comparação de Temperatura nas cidades')
plt.show()
```

```python
import matplotlib.pyplot as plt

hrs_estudadas = [1.5, 3, 5, 7, 4, 2.5]
notas = [6, 8, 10, 10, 3, 3.5]

plt.scatter(hrs_estudadas, notas)
plt.xlabel('Horas Estudadas')
plt.ylabel('Notas')
plt.title('Correlação entre Horas estudadas e Notas')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

cursos = [145, 127, 56, 21, 31, 32]
labels = ['Python', 'R', 'SQL', 'Power BI', 'Excel', 'ChatGpt']
dicionario = {'cursos':cursos, 'labels': labels}
grafico_pizza = pd.DataFrame(dicionario)

plt.pie (x = grafico_pizza.cursos, labels = grafico_pizza.labels)
plt.legend(labels = grafico_pizza.labels, loc = [0.95, 0.35])
plt.show()
```
