Vamos criar uma automação para cadastrar produtos dentro de um sistema.
Os campos a serem preenchidos serão codigo, marca, tipo, categoria, preço unitário, custo e observação.

```python
import pyautogui
import time

pyautogui.PAUSE = 0.5

pyautogui.press("win")

time.sleep(3)

pyautogui.write("chrome")

time.sleep(3)

pyautogui.press("enter")

time.sleep(3)

pyautogui.click(241, 599)

time.sleep(3)

pyautogui.click(1081, 27)

time.sleep(6)

pyautogui.click(356, 61)

time.sleep(2)

pyautogui.write("https://dlp.hashtagtreinamentos.com/python/intensivao/login")

time.sleep(2)

pyautogui.press("enter")

time.sleep(2)

pyautogui.click(427, 405)

time.sleep(2)

pyautogui.write("alunohappybotafogo@outlook.com.br")

time.sleep(2)

pyautogui.click(406, 504)

time.sleep(2)

pyautogui.write("happybotafogo123")

time.sleep(2)

pyautogui.press("enter")
```
