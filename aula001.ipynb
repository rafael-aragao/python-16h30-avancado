```python
import time
import locale

locale.setlocale(locale.LC_TIME, 'pt_BR.UTF-8')

tempo = time.localtime()

print(time.strftime("%d %B %Y", tempo))
print(time.strftime("%H:%M:%S", tempo))
print(time.strftime("%A, %d de %B de %Y, %H:%M:%S", tempo))
```

```python
from datetime import datetime
import locale

locale.setlocale(locale.LC_TIME, 'pt_BR.UTF-8')

agora = datetime.now()

print(agora.strftime("Rio de Janeiro, %A, %d de %B de %Y, %H:%M:%S"))
```

```python
data_hoje = time.ctime()
print(data_hoje)
```
